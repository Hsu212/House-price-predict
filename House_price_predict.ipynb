```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace = True) #This drop NaN values and save the result in the data object again
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
train_data = x_train.join(y_train) #combine the data frame for the training data
train_data
```

```python
train_data.hist(figsize=(15,8)) #figsize is figure size
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap="YlGnBu")
#heatmap is to visualize a correlation Marix & corr() is to produce correlation heat map, cmap is color of map
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1).replace({True:1, False:0}) #to convert categorical variables into numerical dummy variables
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
train_data['bedrooms_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap="YlGnBu")
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
x_train_s = scaler.fit_transform(x_train)

reg = LinearRegression()
reg.fit(x_train_s, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1).replace({True:1, False:0})

test_data['bedrooms_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
x_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
x_test_s = scaler.transform(x_test)
```

```python
reg.score(x_test_s, y_test)
```

Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train_s, y_train)
```

```python
forest.score(x_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV #grid search cross validation

forest = RandomForestRegressor()

param_grid = {
    "n_estimators" : [30,50,100],
    "max_features" : [8,12,20],
    "min_samples_split" : [2,4,6,8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error",return_train_score=True)

grid_search.fit(x_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
best_forest.score(x_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV #grid search cross validation

forest = RandomForestRegressor()

param_grid = {
    "n_estimators" : [100,200,300],
    "min_samples_split" : [2,4],
    "max_depth" : [None,4,8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error",return_train_score=True)

grid_search.fit(x_train_s, y_train)
```
